```python
#create a new dataframe
text_df=pd.DataFrame([open('/content/LAB2.txt')],columns=['Original text'])
text_df
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
#removing Stop Words function
def remove_stopWords(sentence):
    terms=[]
    stopWords= set(stopwords.words('english'))
    for term in sentence.split() :
        if term not in stopWords :
           terms.append(term)
    return " ".join(terms)

text_df["Removed_Stop_word_text"]=text_df["Original text"].apply(remove_stopWords)
print("***************************************************************************documents after removing stopwords*********************************************************************")
text_df
```

```python
import re
import string
#a function to normalize the tweets
def normalize(text_df):
    string = text_df.lower()
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for ele in string:
        if ele in punc:
            string = string.replace(ele, "")
            
    return(string)

text_df["Normalized_text"]=text_df["Removed_Stop_word_text"].apply(normalize)
print("***************************************************************************documents after normalizing*********************************************************************")
text_df
```

```python
#specify that we want to stem arabic text
en_stemmer = stemmer("english")
#define the stemming function
def stem(sentence):
    return " ".join([en_stemmer.stemWord(i) for i in sentence.split()])

text_df['stemmed_text']=text_df['Normalized_text'].apply(stem)
print("***************************************************************************documents after stemming*********************************************************************")
text_df
```
